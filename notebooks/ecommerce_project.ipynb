<h1 style="color: #7B68EE; font-weight: bold; font-family: 'Times New Roman';">Description du Dataset :</h1>

<div style="background-color: #F5F0FF; padding: 25px; border-left: 6px solid #7B68EE; border-radius: 0 15px 15px 0; max-width: 1350px;">

<p style="font-family: 'Times New Roman', Times, serif;">Le projet utilise le dataset <strong style="color: #7B68EE;">Women's E-Commerce Clothing Reviews Dataset</strong>, qui contient des avis laissés par des clientes sur des vêtements vendus sur une plateforme de e-commerce.</p>

<p style="font-family: 'Times New Roman', Times, serif;">Chaque ligne du dataset représente un avis client concernant un produit spécifique. Les données incluent des informations démographiques sur les clientes (comme l'âge), des informations sur le produit (catégorie, département) ainsi que le contenu des avis et la note attribuée.</p>

<p style="font-family: 'Times New Roman', Times, serif; font-weight: bold; color: #7B68EE;">Ces données permettent d'analyser :</p>

<ul style="font-family: 'Times New Roman', Times, serif; list-style-type: square; color: #5a4a9a;">
    <li style="margin-bottom: 5px;">la satisfaction des clientes</li>
    <li style="margin-bottom: 5px;">les caractéristiques des produits les plus appréciés</li>
    <li style="margin-bottom: 5px;">l'impact des avis textuels sur la recommandation d'un produit</li>
    <li style="margin-bottom: 5px;">les relations entre les caractéristiques du produit et la popularité</li>
</ul>

<p style="font-family: 'Times New Roman', Times, serif; font-weight: bold; color: #7B68EE;">Le dataset contient à la fois :</p>

<ul style="font-family: 'Times New Roman', Times, serif; list-style-type: circle; color: #5a4a9a;">
    <li style="margin-bottom: 5px;">des données numériques (âge, rating, feedback count)</li>
    <li style="margin-bottom: 5px;">des données catégorielles (division, department, class)</li>
    <li style="margin-bottom: 5px;">des données textuelles (title, review text)</li>
</ul>

<p style="font-family: 'Times New Roman', Times, serif; background-color: #E6E0FF; padding: 12px; border-radius: 8px; font-style: italic;">Ces différentes variables permettent de réaliser des analyses exploratoires et de construire un modèle de machine learning pour la prédiction de popularité des produits.</p>

</div>

<h1 style="color: #7B68EE; font-weight: bold;font-family: 'Times New Roman';">Objectif de l’analyse :</h1>

<div style="background-color: #F5F0FF; padding: 25px; border-left: 6px solid #7B68EE; border-radius: 0 15px 15px 0; max-width: 1350px;"><p style="font-family: 'Times New Roman', Times, serif;">L'objectif principal de ce projet est de prédire si un produit sera recommandé par les clients ou non, à partir des informations disponibles dans le dataset.</p><p style="font-family: 'Times New Roman', Times, serif; font-weight: bold; color: #7B68EE;">Plus précisément, le projet vise à :</p><ul style="font-family: 'Times New Roman', Times, serif; list-style-type: square; color: #5a4a9a;"> <li style="margin-bottom: 5px;"><strong style="color: #7B68EE;">Comprendre les données clients :</strong> Analyser les caractéristiques des clientes et leur relation avec les avis laissés sur les produits.</li> <li style="margin-bottom: 5px;"><strong style="color: #7B68EE;">Identifier les facteurs de popularité :</strong> Identifier les variables qui influencent la probabilité qu'un produit soit recommandé.</li> <li style="margin-bottom: 5px;"><strong style="color: #7B68EE;">Analyser les avis textuels :</strong> Examiner les commentaires des clientes pour détecter les sentiments positifs ou négatifs.</li> <li style="margin-bottom: 5px;"><strong style="color: #7B68EE;">Construire un modèle de classification :</strong> Développer un modèle de machine learning capable de prédire si un produit sera recommandé ou non.</li> </ul><p style="font-family: 'Times New Roman', Times, serif; font-weight: bold; color: #7B68EE;">Dans ce projet :</p><ul style="font-family: 'Times New Roman', Times, serif; list-style-type: circle; color: #5a4a9a;"> <li style="margin-bottom: 5px;"><strong>Variable cible (Target) :</strong> Recommended IND</li> <li style="margin-bottom: 5px;"><strong>Type de problème :</strong> classification binaire</li> <li style="margin-bottom: 5px;"><strong>Valeurs possibles :</strong></li> <ul style="list-style-type: none; margin-left: 20px;"> <li style="margin-bottom: 3px;">🔵 1 : produit recommandé</li> <li style="margin-bottom: 3px;">⚪ 0 : produit non recommandé</li> </ul> </ul><p style="font-family: 'Times New Roman', Times, serif; background-color: #E6E0FF; padding: 12px; border-radius: 8px; font-style: italic;">L'objectif final est de construire un modèle permettant d'aider une entreprise e-commerce à anticiper la popularité d'un produit en fonction des avis clients et de ses caractéristiques.</p></div>
